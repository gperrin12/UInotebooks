```python
#importing data from a URL - exaple with csv format data
import csv
import requests

#get data from 
#slightly better than just using !curl -O https://serv.cusp.nyu.edu/~fbianco/PUIdata/IntrotoStats_Table6_1.csv

CSV_URL = 'https://serv.cusp.nyu.edu/~fbianco/PUIdata/IntrotoStats_Table6_1.csv'


with requests.Session() as s:
    #downloading
    download = s.get(CSV_URL)

    decoded_content = download.content.decode('utf-8')
    #reading with the csv package requires reading the file, then convering it
    cr = csv.reader(decoded_content.splitlines(), delimiter=',')
    csvData = list(cr)
    
# the syntax above is seen when a process needs to be closed after use. 
# The with...: syntax assures the process is naturally cloased when finished
csvData
```

### JSON file

```python
import sys
import json

try:
    import urllib2 as urllib
except ImportError:
    import urllib.request as urllib


def get_jsonparsed_data(url):
    """
    from http://stackoverflow.com/questions/12965203/how-to-get-json-from-webpage-into-python-script
    Receive the content of ``url``, parse it as JSON and return the object.

    Parameters
    ----------
    url : str

    Returns
    -------
    dict
    """
    response = urllib.urlopen(url)
    data = response.read().decode("utf-8")
    return json.loads(data)
jsonData = get_jsonparsed_data(\
                    "https://serv.cusp.nyu.edu/~fbianco/PUIdata/cityWeather.json")
jsonData
```
