```python
import os
import numpy as np
import pylab as pl
%pylab inline
import pylabsetup
n=np.random.rand(100)*4
print n
```

```python
signs=(np.random.rand(len(n))-0.5)
signs=signs/np.abs(signs)
p = signs*np.sqrt(10**n)/(10**n)*np.random.rand(len(n))
print n.astype(int)*10


fig = pl.figure(1,figsize=(10,5))
ax = fig.add_subplot(111, axisbg='black')
ax.xaxis.grid(True, which="major", linestyle='-', color='white') 
ax.yaxis.grid(True, which="major", linestyle='-', color='white') 
ax.tick_params(color='gray', labelcolor='gray')
for spine in ax.spines.values():
        spine.set_edgecolor('gray')

indx=np.where((p>0.7*signs*np.sqrt(10**n)/(10**n)) * (p > 0))

ax.scatter(10**n,p,c='gray', s=n.astype(int)*100)
ax.scatter(10**n[indx],p[indx],c='r', s=n[indx].astype(int)*100)
ax.set_xscale('log')
ax.set_xlabel("sample size",color='gray')
ax.set_ylabel("effect size",color='gray')

ylims=ax.get_ylim()
xlims=ax.get_xlim()

fig = pl.figure(2,figsize=(10,5))
ax2 = fig.add_subplot(111, axisbg='black')
ax2.xaxis.grid(True, which="major", linestyle='-', color='white') 
ax2.yaxis.grid(True, which="major", linestyle='-', color='white') 
ax2.tick_params(color='gray', labelcolor='gray')
for spine in ax.spines.values():
        spine.set_edgecolor('gray')

ax2.scatter(10**n[indx],p[indx],c='r', s=n[indx].astype(int)*100)

ax2.set_xscale('log')
ax2.set_xlabel("sample size",color='gray')
ax2.set_ylabel("effect size",color='gray')
ax2.set_ylim(ylims)
ax2.set_xlim(xlims)

pl.show()
```
